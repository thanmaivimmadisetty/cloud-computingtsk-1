```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('csvfile.csv')
print(df.head())
df['weather'] = df['weather'].str.strip().str.lower()
weather_map = {
    'rainy': 'rain',
    'raining': 'rain',
    'rain': 'rain',
    'sunny': 'sunny',
    'clear': 'sunny',
    'cloudy': 'cloudy',
    'foggy': 'foggy',
    'snow': 'snow',
    'snowing': 'snow'
}
df['weather'] = df['weather'].map(weather_map).fillna(df['weather'])
df['injury_severity'] = df['injury_severity'].fillna('Unknown')
```

```python
# Q1: Total accidents by location
accidents_by_location = df.groupby('location').size().reset_index(name='total_accidents')
print("Total accidents by location:\n", accidents_by_location)
```

```python
# Q2: Impact of weather on injury_severity
# ----------------------------
severity_by_weather = df.groupby(['weather', 'injury_severity']).size().unstack(fill_value=0)
print("Injury severity by weather:\n", severity_by_weather)

# Optional: percentage distribution
severity_by_weather_percent = severity_by_weather.div(severity_by_weather.sum(axis=1), axis=0) * 100
print("Percentage distribution of injury severity by weather:\n", severity_by_weather_percent)
```

```python
# Q3: Missing injury_severity handled above
# ----------------------------
print("Missing injury_severity replaced with 'Unknown'")
```

```python
# Q4: Correlation between vehicles_involved and severity
# ----------------------------
# First, encode injury_severity numerically
severity_map = {'Minor': 1, 'Moderate': 2, 'Severe': 3, 'Fatal': 4, 'Unknown': 0}
df['injury_severity_numeric'] = df['injury_severity'].map(severity_map)

correlation = df['vehicles_involved'].corr(df['injury_severity_numeric'])
print(f"Correlation between vehicles involved and injury severity: {correlation:.2f}")
```

```python
# Q5: Visualizations
# ----------------------------

# Bar chart: accidents by weather
plt.figure(figsize=(8,5))
sns.countplot(data=df, x='weather', order=df['weather'].value_counts().index, palette='Set2')
plt.title('Number of Accidents by Weather')
plt.xlabel('Weather')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.show()

# Heatmap: severity vs location
# Create pivot table
heatmap_data = df.pivot_table(index='location', columns='injury_severity', aggfunc='size', fill_value=0)

plt.figure(figsize=(12,8))
sns.heatmap(heatmap_data, annot=True, fmt='d', cmap='YlOrRd')
plt.title('Heatmap of Injury Severity vs Location')
plt.xlabel('Injury Severity')
plt.ylabel('Location')
plt.show()
```
